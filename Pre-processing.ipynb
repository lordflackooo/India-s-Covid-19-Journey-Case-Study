```python
import pandas as pd
import numpy as np
```

```python
a2e = pd.read_csv("agedistribution_2016_estimates.csv")
a2e.head()
```

```python
a2e.info()
```

```python
# Print the current column names
print(a2e.columns)

# Rename columns
a2e = a2e.rename(columns={'Age': 'Age_Group', 'M': 'Male', 'F': 'Female', 'Tot': 'Total'})
```

```python
# Replace 'May' with '5' and 'Oct' with '10' in the 'Age' column
a2e['Age_Group'] = a2e['Age_Group'].replace({'May': '5', 'Oct': '10'}, regex=True)
```

```python
a2e.to_csv('Cleaned_agedistribution_2016_estimates.csv', index=False)
```
